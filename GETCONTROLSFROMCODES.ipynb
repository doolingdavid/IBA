## Need to make this process into a function that can loop through all the outsaf files; collect the unique patient ids and accumulate them.

````python
specsdict = {'patient_id': (0,0+11),
            'casereg': (0,0+2),
             'hcpcs': (240,240+5),
             'mf1': (245,245+2),
             'mf2': (247,247+2),
             'mf3': (249,249+2)}

names = sorted(specsdict,key=specsdict.__getitem__)

colspecs = [specsdict[n] for n in names]

import glob
#filelist = glob.glob('./outsaf*')
filelist = glob.glob('outsaf*')
goodfilelist = [f for f in filelist if f.endswith('.txt')]

````

````python
def get_patients_from_file(fin):
    """takes as input the filename, reads it and returns a 
    set containing the patient_id of those who have received PT."""
    
    df = pd.read_fwf(fin, colspecs = colspecs,
                names=names)

    df.patient_id = df.patient_id.astype(str)
    df.patient_id = df.patient_id.apply(lambda x: x.rjust(10, '0'))

    df.set_index('patient_id', inplace=True,drop=False)
    
    
    
    
    good_patients = df.hcpcs.isin(goodcodes) & ~df.hcpcs.isin(treatmentgoodcodes)
    
    
    good_set = set(df.patient_id[good_patients].index)
    
    del df
    
    return good_set
````

## Need to combine the sets returned by the above function.

````python
set0 = get_patients_from_file(goodfilelist[0])
````

````python
set1 = get_patients_from_file(goodfilelist[1])
````

````python
print(set0)
````

````python
print(set1)
````

````python
allsetslist = []
allsetslist.append(set0)
allsetslist.append(set1)

final_result = set.union(*allsetslist)
print(final_result)
````

````python
allsetslist = []

for i in range(len(goodfilelist)):
    s = get_patients_from_file(goodfilelist[i])
    allsetslist.append(s)
    so_far = set.union(*allsetslist)
    print(i, len(list(so_far)))
````

````python
len(allsetslist)
````

````python
import pickle
fileObject = open('control_patient_ids_from_outsaf.pickle', 'wb')
pickle.dump(allsetslist, fileObject)
fileObject.close()
````

````python
fileObject = open('control_patient_ids_from_outsaf.pickle', 'rb')
b = pickle.load(fileObject)
fileObject.close()
allsetslist == b
````
